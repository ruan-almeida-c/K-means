### [Problemas de Clusterização]

A abordagem de clustering pode ser tanto utilizado com métodos supervisionados (quando você tem uma variável categórica com labels conhecidas), quanto pode ser utilizada em um contexto não supervisionado. Um problema de clusterização consiste em analisar e classificar um determinado conjunto de dados, através do descobrimento de padrões grupais, que são descobertos utilizando técnicas que medem o nível de similaridade dos dados analisados.

### [K-Means]

O algoritmo de agrupamento chamado de k-mens( também conhecido como k-médias) é uma técnica de clusterização e caracteriza-se por: não utilizar um supervisor para definir previamente os padrões que serão gerados; ser dependente de uma entidade externa que informe qual a quantidade k de clusters será formado, funciona da seguinte maneira, imagina que você tem uma base de dados na forma:

<img src="https://i.ibb.co/c6TcRJj/dados.png" height="500" width="600">

Veja que os dados estão espalhadas de forma inicialmente irregular, e digamos que sem nenhuma similaridade aparente. Entretanto se eu lhe pedir para dividir esse conjunto de dados em grupos menores, aos quais chamamos de clusters você pode realizar essa tarefa das mais diversas formas possíveis. pois uma das possíveis separações que você fizer, será uma das várias formas que você pode separar os dados, vendo óbviamente somente a imagem anterior como base. Mas qual é o correto? Isso é bem difícil de responder.

O algoritmo de agrupamento __k-means__ dado um grupo de dados com pré conhecido que podemos dividi-los em clusters. O k-means utiliza-se de centroides que seriam os pontos centrais dos grupos que ajudam a encontrar a similaridade entre os dados, o **k** de k-means significa examente esse número.

<img src="https://i.ibb.co/0CPGLZJ/Centroids.png" width="600" height="500">

Como mostrado acime temos 3 centroides distribuidos aleatoriamente juntamente com os dados. A cada centroide será calculado a distância de todos os pontos até aquele centroide, e o ponto será atribuido ao centroide que estiver mais próximo dele.

Na primeira iteração será calculado a distância média de todos os pontos na área designada ao centroide e com isso, irá gerar uma __nova coordenada X e Y__ para onde o centroide irá se deslocar. E logo depois é feito novamente o cálculo das distâncias e atribuído o ponto ao centroide mais próximo.

<img src="https://i.ibb.co/30qt8vc/centroids2.png" width="600">
<img src="https://i.ibb.co/fY5gnn6/centroids2-1.png" width="600">

E com isso será perceptível que a cada iteração os pontos trocaram de centroide.

E terá um momento no qual nenhum ponto mude de centroide e ele encontre um ponto médio no qual terá a distância igualmente entre todos os pontos designado a ele. E com isso teremos os clusters separados.

<img src="https://i.ibb.co/XtYggBX/centroids3.png" width="600">

### [Quantidade de Centroides]

Você deve estar se perguntado "como eu determino o número de centroides", como eu havia mencionado no exemplo anterior, eu já tinha o conhecimento do número de centroides. Agora quando desconhecemos o número de centroides recomendado para um certo grupo de dados, utilizamos um método chamado Método do cotovelo (Ebow Method).


É uma das formas mais utilizadas usadas para descobrir a quantidade ideal de clusters naquele conjunto de dados. O nome se da por ele ter um formato de braço onde o "cotovelo" serve para definir um número estimado de grupos com base nos dados da amostra. Esse método vai aumentando a quantidade de clusters partindo do 1, e analisando o resultado melhorando a cada incremento, podemos observar o momento em que sua performance, o algoritmo busca entrar em um modelo platô no qual a diferença da distância é miníma. Neste momento, é que obtemos o valor para os clusters. #### [Testando o k-means]




<center> Agora, por quê não testamos o k-means para reconhecer esses diferentes grupos, já que sabemos qual seria seus grupos temos aonde comparar se ele está fazendo esse reconhecimento correto ou não.

```python
X = np.array(df.drop("variety", axis = 1))
```

__`df.drop("variety", axis = 1)`__ Na função `drop` excluímos a coluna que indicamos no caso será a `variety` e em seguida indicamos que oque desejamos excluir é a coluna de dados e não a linha, por isso utilizamos o `axis = 1`.

__`np.array()`__ Nessa função do numpy recebemos os dados sem a coluna __"variety"__ e tranformamos num array e armazanamos no `X`.

```python
from sklearn.cluster import KMeans
```

__`from sklean.cluster import KMeans`__ Estamos importando a classe KMeans do sklearn que é uma biblioteca de aprendizado de máquina.

```python
kmeans = KMeans(n_clusters=3,random_state=0)
```

Aqui iniciamos o KMeans com 3 três centroides como especificado em `n_clusters=3` e com suas posições iniciais de forma aleatória mas como queremos repetir algumas vezes com o `random_state=0` podemos diminuir a aleatoriedade das posições dos centroides.

```python
kmeans.fit(X)
```

__`Kmeans.fit(X)`__ A função `fit` faz com que ele "treine" ou melhor dizendo realize as interações até chegar convergir e formar seus grupos com base nos dados contidos em X, que utilizamos anteriormente.

```python
df['k-means'] = kmeans.labels_
```

Podemos criar uma coluna do lado do data frame original para compararmos as respostas recebidas pelo k-means e as repostas que já estão no "variety"

>__Aqui podemos observar a tabela completa com os resultados do k-means e a tabela de variedades do lado.__

```python
df
```

Podemos observar que os resultados encontrados no k-means em alguns pontos.

E se quiser fazer uma comparação de forma visual, podemos gerar o gráfico dos dados gerados pelo k=means e comparar com os gráficos do dataframe original.

```python
sb.pairplot(df,hue="k-means")
```
